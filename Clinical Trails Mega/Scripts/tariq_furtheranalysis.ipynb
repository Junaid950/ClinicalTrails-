```python
import pandas as pd

# Let's create a Dataframe using lists
Months = ['Januaury', 'Feburary', 'March', 'April', 'May',
             'June', 'July', 'August', 'September',
             'October', 'November', 'December']

v2019 = [1368, 1149, 1470,1368, 1342, 1647, 1547, 1406, 1421, 1310,1223, 2690]

v2020 = [1544, 1286, 1740, 1080, 1176,
               1424, 1237, 1126, 1167, 1176,
               1078, 2084]
v2021 =[1131,934,1227,967,984,1094,819,700,528,187,0,0]

# Now, create a pandas dataframe using above lists
df_completion_status = pd.DataFrame(
    {'Months' : Months, '2019' : v2019,
    '2020' : v2020,'2021':v2021})

df_completion_status
```

```python
import matplotlib.pyplot as plt

# Creating figure and axis objects using subplots()
fig, ax = plt.subplots(figsize=[9, 7])

ax.plot(df_completion_status['Months'],
         df_completion_status['2019'],
         marker='o', linewidth=2, label='Completion_Status_2019')
ax.plot(df_completion_status['Months'],
         df_completion_status['2020'],
         marker='o', linewidth=2, 
         label='Completion_Status_2020')
ax.plot(df_completion_status['Months'],
         df_completion_status['2021'],
         marker='o', linewidth=2, 
         label='Completion_status_2021')
plt.xticks(rotation=60)
ax.set_xlabel('Months')
ax.set_ylabel('Year')
plt.legend()
plt.show()
```

```python
import pandas as pd

# Let's create a Dataframe using lists
Types = ['Interventional', 'Observational', 'Ob(Patient-registry)', 'Expanded_access']

v2019 = [255945, 64163, 6171,69]

v2020 = [277631, 71434, 7332, 69]
v2021 =[301472,77540,8180,69]

# Now, create a pandas dataframe using above lists
df_types = pd.DataFrame(
    {'Types' : Types, '2019' : v2019,
    '2020' : v2020,'2021':v2021})

df_types
```

```python
import matplotlib.pyplot as plt

# Creating figure and axis objects using subplots()
fig, ax = plt.subplots(figsize=[9, 7])

ax.plot(df_types['Types'],
         df_types['2019'],
         marker='o', linewidth=2, label='Study_types2019')
ax.plot(df_types['Types'],
         df_types['2020'],
         marker='o', linewidth=2, 
         label='Study_types2020')
ax.plot(df_types['Types'],
         df_types['2021'],
         marker='o', linewidth=2, 
         label='Study_types2021')
plt.xticks(rotation=60)
ax.set_xlabel('Types')
ax.set_ylabel('Year')
plt.legend()
plt.show()
```
